```python
from ase import io
import numpy as np
```

Load in structure to perturb

```python
struct= io.read(filename="./unit_cell/CONTCAR")
struct
```

```python
positions = struct.get_positions()
positions.shape
```

Adjustable percent chance of removing an atom

```python
def make_vacancies(positions, chance=0.03):
    mask = np.random.rand(positions.shape[0])
    mask = np.where(mask < chance, False, True)
    return positions[mask]
```

```python
positions = make_vacancies(positions)
```

Now add Gaussian noise to each position

```python
def add_noise(positions, noise_scale=0.09):
    noise = noise_scale*np.random.randn(*positions.shape)
    return positions+noise
```

```python
add_noise(positions)
```
